```python
from pathlib import Path
import time
import cv2
import shutil
```

```python
MODEL_PATH = "/content/gdrive/MyDrive/Colab Notebooks/doluluk_model_i256-e50-8L.pt"  # Path to trained model file
IMAGE_FOLDER = "/content/gdrive/MyDrive/Colab Notebooks/images" # Folder containing the 1000 images
SAVE_PATH = Path("/content/gdrive/MyDrive/Colab Notebooks/results_speed_test_L")  # Define the save path
SAVE_PATH.mkdir(parents=True, exist_ok=True)  # Create the save path if it does not exist
```

```python
# Load the pretrained YOLO model
model = YOLO(MODEL_PATH)

# Get a list of image files
image_files = list(Path(IMAGE_FOLDER).glob("*.jpg"))[:1000]  # Ensure we have only 1000 images

# Initialize variables to track processing time
start_time = time.time()

# Process each image
for idx, image_path in enumerate(image_files, start=1):
    # Run inference on the image
    results = model(image_path, imgsz=256, conf=0.7)  # Adjust parameters as needed

    # Get the highest confidence class
    label = results[0].names[results[0].probs.top1]  # Get the label name for the top class
    confidence = results[0].probs.top1conf.item()  # Get the confidence score for the top class

    # Save the classification result to a text file
    with open(SAVE_PATH / f"{image_path.stem}_{idx}.txt", "w") as f:
        f.write(f"Label: {label}, Confidence: {confidence:.2f}\n")

    # Save the image with the classification label
    image = cv2.imread(str(image_path))
    cv2.putText(image, f"{label} ({confidence:.2f})", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imwrite(str(SAVE_PATH / f"{image_path.stem}_{idx}.jpg"), image)

# Calculate total processing time
end_time = time.time()
total_time = end_time - start_time

# Print the total processing time
print(f"Total processing time for 1000 images: {total_time:.2f} seconds")
```
